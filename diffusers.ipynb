```python
import torch
from diffusers import StableDiffusion3Pipeline

# Load the model in half-precision (float16) to save memory
pipe = StableDiffusion3Pipeline.from_pretrained(
    "stabilityai/stable-diffusion-3.5-medium",
    torch_dtype=torch.float16
)
pipe.to("cuda")

# Your creative prompt
prompt = "A steampunk robotic owl perched on a crystal branch, digital art, highly detailed, glowing copper gears"

# Generate the image
# Medium models usually need 30-50 steps for peak quality
image = pipe(
    prompt=prompt,
    num_inference_steps=40,
    guidance_scale=7.0
).images[0]

# Display and save
image.save("steampunk_owl.png")
display(image)
```
