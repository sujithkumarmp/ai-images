```python
import torch
from torchvision import models, transforms
from PIL import Image
import requests
import io # Import the io module

# 1. Load a pre-trained Vision Transformer (ViT) - State of the Art!
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.vit_b_16(weights=models.ViT_B_16_Weights.IMAGENET1K_V1).to(device)
model.eval()

# 2. Prepare the image (Resize to 224x224 as required by most models)
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# 3. Grab a sample image (or upload your own)
img_url = "https://images.pexels.com/photos/1108099/pexels-photo-1108099.jpeg" # Changed to a different, reliable JPEG image URL
response = requests.get(img_url)
img = Image.open(io.BytesIO(response.content)) # Modified to use io.BytesIO
img_t = preprocess(img).unsqueeze(0).to(device)

# 4. Run the recognition
with torch.no_grad():
    output = model(img_t)

# 5. Get the result
_, index = torch.max(output, 1)
print(f"Prediction index: {index.item()}")
```
